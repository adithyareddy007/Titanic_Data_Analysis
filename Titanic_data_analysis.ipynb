```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python

data = pd.read_csv('C:/Users/adith/Documents/SCT/train.csv')

```

```python

if 'Survived' not in data.columns:
    raise ValueError("The dataset must contain a 'Survived' column for this analysis.")

```

```python

plt.figure(figsize=(10, 6))
sns.histplot(data['Age'].dropna(), kde=True, bins=30, color='blue')
plt.title('Age Distribution of Passengers', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Sex', hue='Survived', palette='Set2')
plt.title('Gender Distribution by Survival Status', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python


gender_survival = data.groupby(['Sex', 'Survived']).size().unstack()
gender_survival_percentage = (gender_survival.div(gender_survival.sum(axis=1), axis=0) * 100)




gender_survival_percentage.plot(kind='bar', stacked=True, figsize=(10, 6), color=['salmon', 'skyblue'])
plt.title('Survival Percentage by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



print("\nSurvival percentages within each gender:")
print(gender_survival_percentage)


```
